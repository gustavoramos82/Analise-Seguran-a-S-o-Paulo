```python
import pandas as pd
import basedosdados as bd
```

```python
seg = bd.read_table(dataset_id='br_sp_gov_ssp',
table_id='ocorrencias_registradas',
billing_project_id="191635027716")
seg.head()
```

```python
seg = seg[['ano','mes','regiao_ssp','roubo_de_veiculo','roubo_a_banco','roubo_de_carga']]
seg.head()
```

```python
seg['regiao_ssp'].unique()
```

```python
capital = seg[seg['regiao_ssp']=='Capital'].reset_index().drop('index',axis=1)
capital.head()
```

```python
capital.drop('regiao_ssp',axis=1,inplace=True)
capital.head()
```

```python
metropole_sem_capital = seg[seg['regiao_ssp']=='Grande São Paulo (exclui a Capital)'].reset_index().drop('index',axis=1)
metropole_sem_capital.head()
```

```python
metropole_sem_capital.drop('regiao_ssp',axis=1,inplace=True)
metropole_sem_capital.head()
```

```python
outros = seg[(seg['regiao_ssp']!='Capital')&(seg['regiao_ssp']!='Grande São Paulo (exclui a Capital)')].reset_index().drop('index',axis=1)
outros.drop('regiao_ssp',axis=1,inplace=True)
outros.head()
```

```python
capital.to_csv('Capital.csv')
metropole_sem_capital.to_csv('Grande metropole (excluindo capital).csv')
outros.to_csv('Outros.csv')
```
