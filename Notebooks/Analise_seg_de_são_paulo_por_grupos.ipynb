# Olhando em grupos

neste notebook será olhado em grupos, e comparar a análise e verificar se há alguma diferença entre eles

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
capital = pd.read_csv('Capital.csv')
capital.drop('Unnamed: 0',axis=1,inplace=True)
capital.head()
```

```python
metropole = pd.read_csv('Grande metropole (excluindo capital).csv')
metropole.drop('Unnamed: 0',axis=1,inplace=True)
metropole.head()
```

```python
outros = pd.read_csv('Outros.csv')
outros.drop('Unnamed: 0',axis=1,inplace=True)
outros.head()
```

```python
capital_jun = capital.groupby(by='ano').sum()
capital_jun.head()
```

```python
metropole_jun = metropole.groupby(by='ano').sum()
metropole_jun.head()
```

```python
outros_jun = outros.groupby(by='ano').sum()
outros_jun.head()
```

```python
plt.figure(figsize=(15,16))

plt.subplot(3,1,1)

plt.title('Roubo de veículo na capital de São Paulo (2002-2020)',fontsize=14)
sns.lineplot(x='ano',y='roubo_de_veiculo',data=capital_jun)
plt.xlabel('Ano')
plt.ylabel('')

plt.subplot(3,1,2)

plt.title('Roubo de veículo na região metropolitana de São Paulo (2002-2020), exluindo capital',
          fontsize=14)
sns.lineplot(x='ano',y='roubo_de_veiculo',data=metropole_jun)
plt.xlabel('Ano')
plt.ylabel('')

plt.subplot(3,1,3)

plt.title('Roubo de veículo em outras regiões de São Paulo (2002-2020)',fontsize=14)
sns.lineplot(x='ano',y='roubo_de_veiculo',data=outros_jun)
plt.xlabel('Ano')
plt.ylabel('')
```

```python
plt.figure(figsize=(15,16))

plt.subplot(3,1,1)

plt.title('Roubo a banco na capital de São Paulo (2002-2020)',fontsize=14)
sns.lineplot(x='ano',y='roubo_a_banco',data=capital_jun)
plt.xlabel('Ano')
plt.ylabel('')

plt.subplot(3,1,2)

plt.title('Roubo a banco na região metropolitana de São Paulo (2002-2020), exluindo capital',
          fontsize=14)
sns.lineplot(x='ano',y='roubo_a_banco',data=metropole_jun)
plt.xlabel('Ano')
plt.ylabel('')

plt.subplot(3,1,3)

plt.title('Roubo a banco em outras regiões de São Paulo (2002-2020)',fontsize=14)
sns.lineplot(x='ano',y='roubo_a_banco',data=outros_jun)
plt.xlabel('Ano')
plt.ylabel('')
```
